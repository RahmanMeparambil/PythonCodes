### Formatting



strftime() - Convert datetime to string

strptime() - Convert string to datetime

```python
# strftime
dt = datetime.datetime.now()
dt_str = dt.strftime('%B %d, %Y')
print(type(dt_str))
print(dt_str)
```

```python
# strptime
datetime.datetime.strptime(dt_str,'%B %d, %Y')
```
